## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""

# Given data
total_fruits = 100
apples = 60
oranges = 40

##What is the probability that the fruit is apple?

p = apples / total_fruits
print(f"Probability of picking an apple: {p}")

##What is the probability that it is an orange?
q = oranges / total_fruits
print(f"Probability of picking an orange: {q}")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

2. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?
five_apples = p**5
five_apples = round(five_apples, 3)
print(f"Probability of picking the first 5 apples: {five_apples}")

# What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?
p_5_apples_15_oranges = five_apples * (q ** 15)
p_5_apples_15_oranges = round(p_5_apples_15_oranges, 3)
print("Probability of picking 5 apples then 15 oranges:", p_5_apples_15_oranges)
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
from scipy.stats import binom
# Parameters
n = 20       # total fruits picked
k = 5        # number of apples
p = 0.6      # probability of picking an apple

sample_5ap = binom.pmf(k=k, n=n, p=p)
sample_5ap = round(sample_5ap, 3)
print("Probability of the sample caontaining 5 apples and 15 oranges:", sample_5ap)
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here
from scipy.stats import binom
prob_less_than_5 = binom.cdf(k=4, n=n, p=p)

# Round for clarity
prob_less_than_5 = round(prob_less_than_5, 6)

print("P(X < 5) =", prob_less_than_5)
````

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 
n = 20       # total fruits picked
p = 0.6      # probability of apple

# x values: all possible outcomes from 0 to n
x = np.arange(0, n + 1)

# PMF values for each x
pmf = binom.pmf(x, n, p)

# --- Histogram Plot ---
plt.figure(figsize=(8,5))
plt.hist(x, bins=np.arange(-0.5, n+1.5, 1), weights=pmf, color='skyblue', edgecolor='black', rwidth=0.9)

# Labels and title
plt.title('Binomial Distribution - Histogram (n=20, p=0.6)')
plt.xlabel('Number of Apples (k)')
plt.ylabel('Probability P(X = k)')
plt.xticks(range(0, n + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

````python
# your code here 
# Setting the average rate (lambda)
from scipy.stats import poisson
lambda_rate = 2.3  # Average number of events per interval, e.g., 4 arrivals per hour

# Calculating the probability of k events
k_events = 5
prob_k_events = poisson.pmf(k_events, lambda_rate)

print(f"Probability of exactly {k_events} events: {prob_k_events:.4f}")

# Generating a sample of 1000 random variates from the Poisson distribution
sample_poisson = poisson.rvs(mu=lambda_rate, size=100000000)

````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 
# Plotting the sample distribution
plt.hist(sample_poisson, bins=np.arange(sample_poisson.min(), sample_poisson.max()+1)-0.5, density=True, alpha=0.75, color='green')
plt.title('Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.show()
````
